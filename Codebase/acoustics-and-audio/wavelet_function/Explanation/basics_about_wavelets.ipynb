```python
import matplotlib.pyplot as plt       #module for plotting
import pywt                           #open source wavelet transform software
import numpy as np
```

<h2> Different types of continuous wavelet functions</h2>

```python
# Generate a list of available continuous wavelets
wavlist = pywt.wavelist(kind="continuous") 
print(wavlist)
```

<h2> Visualization of different types of continuous wavelet functions </h2>

```python
# Define the number of columns for the plot
cols = 3
# Calculate the number of rows needed based on the number of wavelets and columns
rows = (len(wavlist) + cols - 1) // cols

# Create a subplot grid with the calculated number of rows and columns
fig, axs = plt.subplots(rows, cols, figsize=(10, 10), sharex=True, sharey=True)

# Iterate over the axes and the wavelet list simultaneously
for ax, wavelet in zip(axs.flatten(), wavlist):
    # Adjust wavelet names that require parameters
    if wavelet in ['cmor', 'shan']:
        wavelet += '1-1'
    elif wavelet == 'fbsp':
        wavelet += '1-1.5-1.0'

    # Generate the wavelet function (psi) and its x-coordinates for plotting
    [psi, x] = pywt.ContinuousWavelet(wavelet).wavefun(10)
    
    # Plot the real part of the wavelet function
    ax.plot(x, np.real(psi), label="real")
    # Plot the imaginary part of the wavelet function
    ax.plot(x, np.imag(psi), label="imag")
    # Set the title of the subplot to the wavelet name
    ax.set_title(wavelet)
    # Set the x-axis limits
    ax.set_xlim([-5, 5])
    # Set the y-axis limits
    ax.set_ylim([-0.8, 1])

# Add a legend to the last axis in the grid
ax.legend(loc="upper right")

# Add a main title to the entire figure
plt.suptitle("Available wavelets for CWT")

# Adjust the layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
