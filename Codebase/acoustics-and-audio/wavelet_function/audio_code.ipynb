```python
import matplotlib.pyplot as plt       #module for plotting
import numpy as np 
import pywt                           #open source wavelet transform software
import librosa                        # for audio files
from scipy.signal import convolve2d   # performs convolution
from scipy.io.wavfile import write    # save audio file als wav file
```

<h2> Generate an Audio Signal </h2>

```python
# Parameters for the audio signal
duration = 2.0  # seconds
sampling_rate = 44100  # samples per second
frequency = 440.0  # frequency of the sine wave (A4 note)
amplitude = 0.5 # scale the wave
# Generate the time axis
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate the sine wave
signal_selfmade = amplitude * np.sin(2 * np.pi * frequency * t)

# Save the audio signal as a WAV file
audio_filename = 'sine_wave.wav'
write(audio_filename, sampling_rate, signal_selfmade.astype(np.float32))
```

<h2> Import/Load an Audio Signal </h2>

```python
# Load audio file
def signal_load():
    audio_path = 'audio_signal.wav'  # Replace with your audio file path
    signal, sr = librosa.load(audio_path, sr=None)
    plt.plot(signal) # plot the audio signal 
    plt.xlabel("sample rate") #sample rate
    plt.ylabel("elongation") #voltage, elongation
    print(sr)
    return signal, sr
signal, sr = signal_load()
print(signal)
```
