This notebook presents basic and easily customisable functions for additive and substractive synthesis, as well as for frequency modulation. Even though chaining all the presented functions is possible through rearranging and modifying the relevant lines, or even through a clever use of certain parameters, creating a more convenient method that can utilise all the functions at once is the obvious next step in the development of the code.

For the actual part of the code, NumPy will suffice. SciPy might be a lot more convenient to use for signal processing, but we also won't be needing anything more than basic trigonometric functions, some constants, and the forward and the inverse Fourier transforms. Matplotlib and SciPy's Wavfile Write library are also imported for the purposes of visualising and listening to our data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

We start by defining our wave forms under a single function, so they can be called when necessary.

```python
fsample = 44100
samplerate = 1 / fsample

def signal(f, waveform = "sin", T = 1, A = 1, phi = 1):

    t = np.arange(0, T, samplerate)
    
    if waveform == "sin":
        wave = A * np.cos(2 * np.pi * f * t - 2 * np.pi * phi)
    if waveform == "squ":
        wave = A * np.sign(np.cos(2 * np.pi * f * t - 2 * np.pi * phi))
    if waveform == "tri":
        wave = 2 * abs(A - 2 * A * ((t - (phi / f)) % (1 / f)) * f) - A
    if waveform == "saw":
        wave = A - 2 * A * ((t - (phi / f))  % (1 / f)) * f
    if waveform == "saw_rise":
        wave = -(A - 2 * A * ((t - (phi / f))  % (1 / f)) * f)
    if waveform == "rnd":
        wave = np.random.normal(0, (np.exp(-0.5 / (A**2))) / (A * np.sqrt(2 * np.pi)), len(t))
    return wave
```

Additive synthesis is the most straightforward method, since it basically returns a single signal resulting from the sum of multiple signals. However, we should also have the means to control the parameters of all those signals. Below is an example that uses some arbitrary for loops to manipulate amplitude, phase shift and frequency of the added signals in relation to the "fundamental" signal.

```python
def addit(f, K, f_n = 1, A_n = 0.02, phi_n = 0.02, waveform = "sin", T = 1, A = 1, phi = 1):

    t = np.arange(0, T, samplerate)
    partials = []

    for N in range(K):
        F = f + f * N * f_n
        B = A - A_n * N
        Phi = phi - phi_n * N
        partial = (1 / K) * signal(F, waveform, T, B, Phi)
        partials.append(partial)

    return sum(partials)
```

We could even modify the function to use custom lists. Below is an example that omits certain frequencies while still following a pattern like the overtone series. The same could be done for the other parameters as well, including the waveform.

```python
my_partials = []

def addit_custom1(f, f_n = 1, A_n = 0.02, phi_n = 0.02, waveform = "sin", T = 1, A = 1, phi = 1):

    t = np.arange(0, T, samplerate)
    partials = []

    for N in range(len(my_partials)):
        F = f + f * my_partials[N] * f_n
        B = A - A_n * N
        Phi = phi - phi_n * N
        partial = (1 / len(my_partials)) * signal(F, waveform, T, B, Phi)
        partials.append(partial)

    return sum(partials)
```

```python
addit1 = addit(16, 5, A_n = 0.1, phi_n = 0)
addit2 = addit(16, 5, A_n = 0.1, phi_n = 0, f_n = 0.9)

my_partials = [1, 0, 0, 1, 0, 0]
addit3 = addit_custom1(16)

t = np.arange(0, 1, samplerate)
plt.subplot(3, 1, 1)
plt.plot(addit1)
plt.subplot(3, 1, 2)
plt.plot(addit2)
plt.subplot(3, 1, 3)
plt.plot(addit3)
plt.show()

#convert to audible frequencies
addit10 = addit(256, 5, A_n = 0.1, phi_n = 0)
addit20 = addit(256, 5, A_n = 0.1, phi_n = 0, f_n = 0.9)
addit30 = addit_custom1(256)

write("pymus_addit1.wav", fsample, addit10.astype(np.float32))
write("pymus_addit2.wav", fsample, addit20.astype(np.float32))
write("pymus_addit3.wav", fsample, addit30.astype(np.float32))
```

The FM synthesis is also relatively easy to compute. In frequency modulation, there is a carrier signal and a message signal, both with their own base frequencies. The frequency of the carrier frequency, however, increases or decreases depending on the message signal. This change can be represented as an additional signal, like this:

$$ sin( 2\pi f_{c} t + sin(2\pi f_{m} t) )$$

The exact equation for the frequency modulation requires additional steps, however this simplified equation already implies that the message signal can be written similar to the phase shift parameter. Frequency modulation is a type of angle modulation after all. That said, we need to consider that phase modulation is also a method which is similar to frequency modulation. To avoid phase modulation, we need to introduce what is called a modulation index into our function. Then, after shifting the message signal such that it looks more like a sine wave instead of a cosine wave, we can write our function as follows:

```python
def fm(f, f_m, waveform = "sin", waveform_m = "sin", T = 1, A = 1, A_m = 100, phi = 1, phi_m = 1):

    t = np.arange(0, T, samplerate)
    
    if not waveform_m == "rnd":
        wave = signal(f, waveform, T, A, phi + (-1 / (2 * np.pi * f_m)) * signal(f_m, waveform_m, T, A_m, phi_m + 0.25))

    if waveform_m == "rnd":
        wave = signal(f, waveform, T, A, phi + (-1 / (2 * np.pi)) * signal(1, waveform_m, T, A_m, phi_m + 0.25))

    return wave
```

Our carrier signal is defined as a regular signal with the exception of the phase shift parameter. We add the message signal to the phase shift with the modulation index as its coefficient, and start its phase from $\pi$. This is practical, because we defined our sawtooth and our triangle waves without using a trigonometric function. We also need to consider the noise wave, so we add an if statement in which we set the frequency of the noise message to 1 to avoid unintentionally reducing the modulation index.

```python
fm1 = fm(128, 64, T = 0.04)
fmA = signal(64, T = 0.04)
fm2 = fm(128, 127, T = 0.04)
fmB = signal(125, T = 0.04)
fm3 = fm(128, 64, "saw", "squ", T = 0.04)
fmC = signal(64, "squ", T = 0.04)
fm4 = fm(128, 64, "tri", "rnd", T = 0.04, A_m = 10)
fmD = signal(64, "rnd", T = 0.04, A = np.e)

plt.subplot(2,2,1)
plt.plot(fmA)
plt.plot(fm1, "r")
plt.subplot(2,2,2)
plt.plot(fmB)
plt.plot(fm2, "r")
plt.subplot(2,2,3)
plt.plot(fmC)
plt.plot(fm3, "r")
plt.subplot(2,2,4)
plt.plot(fmD)
plt.plot(fm4, "r")
plt.show()
```

An important thing to note is that this FM synthesis function can also be written using the regular signal function. This also means that we can chain multiple frequency modulations in a single line to create a linear FM synthesis.

```python
linear1 = signal(1.2, waveform = "saw", T = 0.5)
linear2 = signal(12.8, waveform = "squ", T = 0.5)
linear3 = signal(25.6, waveform = "sin", T = 0.5)
linear_fm = signal(25.6, waveform = "sin", T = 0.5, phi = 1 +  (-1 / (2 * np.pi * 12.8)) * signal(12.8, waveform = "squ", T = 0.5, A = 50, phi = 0.25 + 1 + (-1 / (2 * np.pi * 1.2)) * signal(1.2, waveform = "saw", T = 0.5, A = 100, phi = 0.25)))

linear10 = signal(12, waveform = "saw", T = 5)
linear20 = signal(128, waveform = "squ", T = 5)
linear30 = signal(256, waveform = "sin", T = 5)
linear_fm0 = signal(256, waveform = "sin", T = 5, phi = 1 +  (-1 / (2 * np.pi * 128)) * signal(128, waveform = "squ", T = 5, A = 50, phi = 0.25 + 1 + (-1 / (2 * np.pi * 12)) * signal(12, waveform = "saw", T = 5, A = 100, phi = 0.25)))

plt.subplot(2,2,1)
plt.plot(linear1)
plt.subplot(2,2,2)
plt.plot(linear2)
plt.subplot(2,2,3)
plt.plot(linear3)
plt.subplot(2,2,4)
plt.plot(linear_fm)
plt.show()

write("pymus_linear1.wav", fsample, linear10.astype(np.float32))
write("pymus_linear2.wav", fsample, linear20.astype(np.float32))
write("pymus_linear3.wav", fsample, linear30.astype(np.float32))
write("pymus_linear_fm.wav", fsample, linear_fm0.astype(np.float32))
```

In this notebook, there are two functions for the substractive synthesis. The first one is relatively simple. We first apply Fourier transformation to a signal and get a frequency domain. We then define an arbitrary filter function that is symetrically distributed over the frequency domain. Lastly, we apply the filter by multiplying the filter function with the frequency domain and by applying the inverse Fourier transformation to get the filtered signal.

```python
def sub(f, waveform, T = 1, A = 1, phi = 1, k = 1, x0 = 0, L = 1):

    t = np.arange(0, T, samplerate)

    filter_pos = []
    filter_neg = []
    filter_sum = [0]

    wave_sub = signal(f, waveform, T, A, phi)
    f_domain = np.fft.fft(wave_sub)

    for N in range(int(fsample / 2)):
        logistic_pos = L / (1 + np.e**(-12 * k * samplerate * (N - x0 - fsample / 4)))
        logistic_neg = L / (1 + np.e**(-12 * k * samplerate * ((fsample / 2 - 1 - N) - x0 - fsample / 4)))
        filter_pos.append(logistic_pos)
        filter_neg.append(logistic_neg)

    del filter_pos[-1]
    filter_sum.extend(filter_pos)
    filter_sum.extend(filter_neg)    

    f_domain_filtered = f_domain * filter_sum
    wave = np.fft.ifft(f_domain_filtered)

    return wave
```

Since the frequency domain starts with 0 Hz, we can set our first filter parameter to 0 as well. Then, we will need to adjust the left side of our filter such that the distribution becomes symmetrical. For the above example, this is as simple as removing the last element of the left side list since that value would correspond to the first value of the right side list. This is much better visualised than explained.

```python
wave = signal(100, "squ")
f_domain = np.fft.fft(wave)

L = 1 #maximum
k = -2 #steepness, negative for lowpass
x0 = 1000 #first value where something happens

filter_pos = []
filter_neg = []
filter_sum = [0]

for N in range(int(fsample / 2)):
    logistic_pos = L / (1 + np.e**(-12 * k * samplerate * (N - x0 - fsample / 4)))
    logistic_neg = L / (1 + np.e**(-12 * k * samplerate * ((fsample / 2 - 1 - N) - x0 - fsample / 4)))
    filter_pos.append(logistic_pos)
    filter_neg.append(logistic_neg)

del filter_pos[-1]
filter_sum.extend(filter_pos)
filter_sum.extend(filter_neg)

f_domain_filtered = f_domain * filter_sum
wave_out = np.fft.ifft(f_domain_filtered)

plt.subplot(3, 2, 1)
plt.xlim(0, 2205)
plt.plot(wave)

plt.subplot(3, 2, 2)
plt.plot(f_domain)

plt.subplot(3, 2, 3)
plt.xlim(0, 2205)
plt.plot(wave_out)

plt.subplot(3, 2, 4)
plt.plot(f_domain_filtered)

plt.subplot(3, 2, 5)
plt.plot(filter_pos)

plt.subplot(3, 2, 5)
plt.plot(filter_neg)

plt.subplot(3, 2, 6)
plt.plot(filter_sum)

plt.show()
```

The second substractive synthesis function is a little more performance intensive, so we need to avoid values that are too large. However, it can come in as a very handy tool for synthesizing realistic signals. Our computers process information with extreme precision, but in the real world an electrical signal transmission can have barely noticable changes on the signal itself. A simple substractive synthesis is enough to replicate the changes on the amplitude of the overtones, but the phase of the overtones also get "filtered" as a signal is transmitted. This advanced substractive synthesis aims to create filters that can alter the amplitudes as well as the phases of the overtones.

Our first goal in this filter is to normalize the frequency parameters from the Fourier transform in such a way where they become comprehensible. We do this by converting those values, which are complex numbers, to their exponential form. We can now take each value from our new frequency domain and put them directly into our own manual inverse Fourier transformation. Since our values are complex, we need to create both a sine-like and a cosine-like transformation using real and imaginary components respectively. The sawtooth wave, for example, can only be transformed back using the imaginary components, while our other waveforms can only be transformed back using the real components.

```python
def sub_adv(f, waveform, phi_post_rate = 0, A_post_rate = 0, A_post_steep = 0, T = 1, A = 1, phi = 1, f_sub = 4):
    
    t = np.arange(0, T, samplerate)
    wave_sub = signal(f_sub, waveform, 1 / f_sub, A, phi)
    f_domain = np.fft.fft(wave_sub)
    f_domain_len = int((len(f_domain) - 1) / 2)
    f_domain_post = []
    wave_sub_post = []
    phi_post = []
    A_post = []

    for N in range(f_domain_len):
        phi_post_val = phi_post_rate * (N / f_domain_len)
        phi_post.append(phi_post_val)
    
    for N in range(f_domain_len):
        A_post_val = 1 - (A_post_rate * ((N / f_domain_len)**A_post_steep))
        A_post.append(A_post_val)
        
    
    for N in range(f_domain_len):
        f_domain_normalize = 2 * f_sub * np.abs(f_domain[N+1] * samplerate) * np.exp(1j * np.angle(f_domain[N+1]))
        f_domain_post.append(f_domain_normalize)

    for N in range(1, f_domain_len + 1):
      inv_transform_pos = (np.real(f_domain_post[N-1]) * A_post[N-1]) * np.cos(2 * np.pi * N * f * t - 2 * np.pi * phi_post[N-1])
      wave_sub_post.append(inv_transform_pos)
      inv_transform_neg = (np.imag(f_domain_post[N-1]) * A_post[N-1]) * np.cos(2 * np.pi * N * f * t - 2 * np.pi * phi_post[N-1] + 0.5 * np.pi)
      wave_sub_post.append(inv_transform_neg)

    return sum(wave_sub_post)
```

As it was the case with the previous functions, the for loops that alter the amplitude and the phase shift are arbitrarily defined and should be changed if necessary. Another thing worth noting is that for unknown reasons, this function works exceptionally well for signals with 4 Hz in terms of accuracy and performance, therefore all the operations until the inverse transformation assume that the original signal is at 4 Hz. The frequency is only corrected during the inverse transformation and the white noise wave can't be used with this filter. It is not recommended to change f_sub other than testing purposes.

```python
subadvA = signal(128, "squ")
subadv1 = sub_adv(128, "squ", 100, 1, 0.5)
subadvB = signal(128, "saw")
subadv2 = sub_adv(128, "saw", 0.5, 2, 0.2)
```

```python
plt.subplot(2,2,1)
plt.xlim(0, 2205)
plt.plot(subadvA)
plt.subplot(2,2,3)
plt.xlim(0, 2205)
plt.plot(subadv1)
plt.subplot(2,2,2)
plt.xlim(0, 2205)
plt.plot(subadvB)
plt.subplot(2,2,4)
plt.xlim(0, 2205)
plt.plot(subadv2)
plt.show()

write("pymus_subadv1_unfiltered.wav", fsample, subadvA.astype(np.float32))
write("pymus_subadv1_filtered.wav", fsample, subadv1.astype(np.float32))
write("pymus_subadv2_unfiltered.wav", fsample, subadvB.astype(np.float32))
write("pymus_subadv2_filtered.wav", fsample, subadv2.astype(np.float32))
```
