<h3>Sinuskurve plotten try 2</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

fsample = 44100
samplerate = 1/fsample

def signal(f, T = 1, A = 1, phi = 1, wavcycles = 2, filename='output.wav'):

    
    t = np.arange(0, T, samplerate)
    
    wave = A * np.sin(2 * np.pi * f * t - 2 * np.pi * phi)
    # A steht für Amplitüde, also die maximale Auslenkung von der Ruhelage (y=0). In diesem Fall ist A=1 die maximale Lautstärke, 
    #die von der Welle erreicht werden kann. 
    # Wir brauchen Numpy auch um die Sinusfunktion hinzuzufügen, in der Klammer spezifizieren wir im ersten Teil (?) und im zweiten Teil
    #die Phasenverschiebung, 2pi ist genau ein Zyklus, deswegen repräsentiert phi als Faktor die Phasenverschiebung
    # 
    
     # Plotting
    
    plt.figure(figsize=(10, 6))
    plt.plot(t, wave)
    plt.title(f'Signal: frequency={f} Hz, amplitude={A}, phase shift (φ)={phi}')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.xlim(0, (T/T)/(f/wavcycles)) # hier ist wavcycles wichtig
    plt.ylim(-A-0.5, A+0.5)
    plt.axhline(0, color='black', lw=0.5, ls='--')  # diese axis symbolisiert A=0
    plt.show()  # Show the plot
    
    wave_int16 = np.int16(wave * 32767)
    write('sine_wave.wav', fsample, wave_int16)

signal(f=4000, T=1, A=1, phi=1, wavcycles = 4, filename='sinus.wav')






    
    
    
```

<h3>Try 3</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write

samplerate = 44100

def signal(f=440, T = 1, A = 1, phi = 1, wavcycles = 1, filename='output.wav'):

    
    t = np.linspace(0, T, int(T*samplerate), endpoint=False)
    
    wave = A * np.sin(2 * np.pi * f * t - 2 * np.pi * phi)

    wave_int16 = np.int16(wave * 32767)
    write('sine_wave.wav', samplerate, wave_int16)
    
    
    plt.figure(figsize=(10, 6))
    plt.plot(t, wave)
    plt.title(f'Signal: Frequenz={f} Hz, Amplitüde={A}, Phasenverschiebung (φ)={phi}')
    plt.xlabel('Zeit (s)')
    plt.ylabel('Amplitüde')
    plt.grid()
    plt.xlim(0, 1/(f/wavcycles)) 
    plt.ylim(-A-0.5, A+0.5)
    plt.axhline(0, color='black', lw=0.5, ls='--') 
    plt.show()  # Show the plot


signal(f=440, T=1, A=0.8, phi=0, wavcycles = 2, filename='sinus.wav')





    
    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

samplerate = 44100

def signal(f=440, T = 1, A = 1, phi = 1, wavcycles = 1):
    
    t = np.linspace(0, T, int(T*samplerate), endpoint=False)
    wave = A * np.sin(2 * np.pi * f * t - 2 * np.pi * phi)

    plt.figure(figsize=(10, 6))
    plt.plot(t, wave)
    plt.title(f'Signal: Frequenz={f} Hz, Amplitüde={A}, Phasenverschiebung (φ)={phi}')
    plt.xlabel('Zeit (s)')
    plt.ylabel('Amplitüde')
    plt.grid()
    plt.xlim(0, 1/(f/wavcycles)) 
    plt.ylim(-A-0.5, A+0.5)
    plt.axhline(0, color='black', lw=0.5, ls='--') 
    plt.show()  # Show the plot

signal(f=440, T=1, A=0.8, phi=0, wavcycles = 2)









```
