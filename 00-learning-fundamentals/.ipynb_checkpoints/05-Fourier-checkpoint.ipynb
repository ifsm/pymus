<h1>Fourier Details</h1>
Discrete Fourier transforms tend to cause numerical errors known as leakage effect. The effect cannot be eliminated. But the right windowing function can reduce the effect a bit. Famous windowing methods are Blackman, Blackman-Harris, Hann, Hamm, Bartlett, Bartlett-Hann, Bohman, Dolph-Chebyshev, Cosine, Kaiser, Gauss. A windowing method is nothing but a temporal envelope function, i.e., a matrix multiplication <i>A Ã— B = C</i>, where <i>A = [-1, -.99, ... .54]</i> is the time series to analyze (signal), <i>B = [0, ..., 1, ..., 0]</i><sup>T</sup> is the windowing function and <i>C = [0, ..., 0]</i> is the windowed signal.

```python
from scipy.fft import fft, ifft#, fftfreq
from scipy.signal import blackman
from scipy.signal import blackmanharris
from scipy.signal import bartlett
from scipy.signal import hann
from scipy.signal import hamming
from scipy.signal import kaiser
from scipy.signal import gaussian
from scipy.signal import bohman
import matplotlib.pyplot as plt # for plots
import numpy as np#includes mathematical functions, like log10(), abs(), and constants, like pi
```

Some exemplary windowing functions are shown below. They are included in the scipy library. They are symmetric curves that start near or at <i>0</i>, peak at <i>1</i> and return to the initial value of/near <i>0</i>:

```python
fs = 512 # frame size of the signal to be analyzed
sr = 8000 # sample rate
# some windowing functions
blackmanharriswindow = blackmanharris(fs)
blackmanwindow = blackman(fs)
bartlettwindow = bartlett(fs)
hannwindow = hann(fs)
hammwindow = hamming(fs)
kaiserwindow = kaiser(fs,4)

plt.plot(range(fs),blackmanharriswindow)# blue
plt.plot(range(fs),blackmanwindow)# orange
plt.plot(range(fs),bartlettwindow)# green
plt.plot(range(fs),hannwindow)# red
plt.plot(range(fs),hammwindow)# violet
plt.plot(range(fs),kaiserwindow)# brown
```

These curves are the windowing functions that we referred to as <i>B</i>. Let us now define and plot a signal <i>A</i>:

```python
time=np.arange(0,fs/sr,1/sr)
signal = np.cos(2*np.pi*50*time)+np.cos(2*np.pi*100*time)
plt.plot(signal)
```

This signal <i>A</i> is made of a <i>50</i> Hz frequency and a <i>100</i> Hz frequency. Both have the same amplitude. What does the frequency spectrum look like?

```python
xfreq = np.arange(0,sr//2,sr/fs)# assigns each frequency bin its frequency
spec = fft(signal)# Calculate the frequency spectrum of the signal A
magspec=np.abs(spec)# isolates the magnitude (ignoring the phase)
magspec=magspec[0:magspec.size //2]# eliminate negative frequencies
specdB = 20 * np.log10(magspec / max(magspec))# provide amplitudes in dB
plt.plot(xfreq,specdB)#plot amplitudes over frequency
plt.grid()
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("f [Hz]", fontsize = 16)
plt.ylabel("p [dB]", fontsize = 16)
plt.xlim(1,300)# range to 300 to zoom in
plt.ylim(-60,0)
```

```python
print(specdB[2])
print(specdB[3])
print(specdB[4])
print(specdB[5])
print(specdB[6])
print(specdB[7])
```

The plot looks partly as expected:
<ul>
    <li>We see peaks at <i>50</i> Hz and <i>100</i> Hz</li>
    <li>The neighboring frequency bins have amplitudes of <i>-13, -17, -19</i> and <i>-5</i> dB</li>
    <li>There are no other protruding frequency bins</li>
</ul>
But some observations may seem unexpected:
<ul>
    <li><i>50</i> Hz and <i>100</i> Hz do not have the same amplitude, but differ by <i>3.5</i> dB</li>
    <li>All frequencies below <i>400</i> Hz have an amplitude <i>> -30</i> dB</li>
    <li>All frequencies have an amplitude larger than <i>-50</i> dB</li>
</ul>
Why is that so? This is because the Fourier transform expects a signal to start at <i>t<sub>0</sub> = - âˆž</i> and end at <i>t<sub>N</sub> = âˆž</i>. But our signal has a finite length. We analyze the signal from sample number <i>0</i> to sample number <i>511</i>. This way, we implicitly do as if we analyzed one period from a signal that repeats forever. Such a signal would exhibit a sudden (steep) "jump" from <i>-0.4142889085225397</i> to <i>2</i> every period, since our signal does not end where it started (i.e., the signal's actual period is not an integer multiple of the frame size). Under this assumption, the frequency spectrum is indeed correct. Signal theorists know that steep jumps are caused by many frequencies in phase. This is what we see: All frequencies are contained according to out Fourier transform. Their amplitudes range from <i>0</i> to <i>-46.5</i> dB. But what if the analyzed signal is not one period of an ever-repeating signal? I think what we mean when we define a signal like this is that two frequencies with equal amplitudes last for a while. And we would like the Fourier analysis to display these two frequencies, and not the rest. In other words: We consider the signal as containing only two frequencies. What the plot displays is called a leakage effect. The energy from the two frequencies leak into neighboring frequencies, as a consequence of our imperfect constellation where we analyze some random excerpt of a signal and not one period. To reduce the leakage effect and retrieve the two expected frequencies, each with the same amplitude, we can apply a windowing function BEFORE carrying out the Fourier transform. Here is what windowed signals look like:

```python
bhapplied = blackmanharriswindow * signal/2
plt.plot(bhapplied)
plt.plot(blackmanharriswindow)
```

```python
bapplied = blackmanwindow * signal/2
plt.plot(bapplied)
plt.plot(blackmanwindow)
```

```python
bartapplied = bartlettwindow * signal/2
plt.plot(bartapplied)
plt.plot(bartlettwindow)
```

```python
hannapplied = hannwindow * signal/2
plt.plot(hannapplied)
plt.plot(hannwindow)
```

```python
hammapplied = hammwindow * signal/2
plt.plot(hammapplied)
plt.plot(hammwindow)
```

```python
kaiserapplied = kaiserwindow * signal/2
plt.plot(kaiserapplied)
plt.plot(kaiserwindow)
```

```python
spec = fft(bhapplied)# Calculate the frequency spectrum of the signal A
magspec=np.abs(spec)# isolates the magnitude (ignoring the phase)
magspec=magspec[0:magspec.size //2]# eliminate negative frequencies
specdB = 20 * np.log10(magspec / max(magspec))# provide amplitudes in dB
plt.plot(xfreq,specdB)#plot amplitudes over frequency
plt.grid()
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("f [Hz]", fontsize = 16)
plt.ylabel("p [dB]", fontsize = 16)
plt.xlim(1,300)# range to 300 to zoom in
plt.ylim(-60,0)
```

```python
print(specdB[2])
print(specdB[3])
print(specdB[4])
print(specdB[5])
print(specdB[6])
print(specdB[7])
```

What effect had the Blackm-Harris window?
<ul>
    <li>The time series changed</li>
    <li>The time series starts and ends at <i>0</i> eliminating the "jump"</li>
    <li>Details from the beginning and the end of the time frame are lost</li>
    <li>The windowing function (envelope) is somewhat visible in the time series</li>
    <li>The spectrum has changed</li>
    <li>The two peaks from before are still clearly visible</li>
    <li>No frequency above <i>150</i> Hz has a mentionable amplitude</li>
    <li><i>50</i> Hz and <i>100</i> Hz have a more similar amplitude now</li>
    <li>But: The direct neighbors have a much higher amplitude than before: <i>-4.7, -2.8, -10.7</i> and <i>-1</i> dB</li>
</ul>

This is a typical result of windowing: The energy of contained frequencies leaks more into its direct neighbors, but much less into other frewuency bins. In other words: The main benefit of windowing is that energy of frequencies does not leak into higher frequency bins that much. The main drawbacks are that windowing hides what is contained in the beginning and at the end of the time frame, and the width of the peaks increases, i.e., a steady frequency looks like an amplitude-modulated frequency (beats), because windowing produces these beats.

Every windowing function behaves a little different. The Kaiser window, for example, has much less leakage into neighboring frequency bins, but more into the rest of the spectrum.

```python
spec = fft(kaiserapplied)# Calculate the frequency spectrum of the signal A
magspec=np.abs(spec)# isolates the magnitude (ignoring the phase)
magspec=magspec[0:magspec.size //2]# eliminate negative frequencies
specdB = 20 * np.log10(magspec / max(magspec))# provide amplitudes in dB
plt.plot(xfreq,specdB)#plot amplitudes over frequency
plt.grid()
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("f [Hz]", fontsize = 16)
plt.ylabel("p [dB]", fontsize = 16)
plt.xlim(1,300)# range to 300 to zoom in
plt.ylim(-60,0)
```
